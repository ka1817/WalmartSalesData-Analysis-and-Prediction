# Train Test Split

```python
# Define X and y
X = df.drop(columns=['Invoice ID', 'Date', 'Time', 'Total','Branch'])
y = df['Total']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Training Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

# Model Evaluation

```python
predictions=lm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

# Example single input data

```python
import pandas as pd
data = {
    'City': ['Yangon'],
    'Customer type': ['Normal'],
    'Gender': ['Female'],
    'Product line': ['Health and beauty'],
    'Unit price': [50.0],
    'Quantity': [10],
    'Tax 5%': [25.0],
    'Total': [525.0],  # This is what we want to predict
    'Date': ['2020-05-01'],
    'Time': ['12:00:00'],
    'Payment': ['Ewallet'],
    'cogs': [500.0],
    'gross margin percentage': [4.761905],
    'gross income': [25.0],
    'Rating': [8.0],
    'Month': ['May'],
    'Year': [2020]
}

# Convert to DataFrame
df_single = pd.DataFrame(data)

# Convert 'Date' column to datetime
df_single['Date'] = pd.to_datetime(df_single['Date'])
df_single['Month'] = df_single['Date'].dt.strftime('%B')
df_single['Year'] = df_single['Date'].dt.year

# Drop 'Date' and 'Time' columns if not used in model
df_single = df_single.drop(columns=['Date', 'Time'])
```

```python
# Apply one-hot encoding to categorical columns
df_single_encoded = pd.get_dummies(df_single, columns=['City', 'Customer type', 'Gender', 'Product line', 'Payment', 'Month'], drop_first=True)
```

```python
# Assuming `X` is the DataFrame used for training and has the columns in the correct order
# Align the columns in the single input DataFrame
for col in X.columns:
    if col not in df_single_encoded.columns:
        df_single_encoded[col] = 0

# Reorder columns to match the training data
df_single_encoded = df_single_encoded[X.columns]
```

```python
# Use the trained model to predict the value
single_prediction = lm.predict(df_single_encoded)

# Output the predicted value
print(f'Predicted Total: {single_prediction[0]}')
```
